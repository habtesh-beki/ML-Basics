```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
data = {
    'Emproye id':[10,20,15,25,30],
    'Gender':['M','F','F','M','F'],
    'Remarks':['Good','Nice','Good','Great','Nice']
}
df = pd.DataFrame(data)
print(df)
```

```python
categorical_column = df.select_dtypes(include=['object']).columns.tolist()

encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[categorical_column])
one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_column))
df_encoded = pd.concat([df,one_hot_df],axis=1)

df_encoded = df_encoded.drop(categorical_column,axis=1)
print(f"Encoded employs data: \n {df_encoded}")
```
