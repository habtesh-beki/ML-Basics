```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
```

```python
dog_whisker_length = np.random.normal(loc=5, scale=1 , size=10 )
dog_ear_flapiness_index = np.random.normal(loc=8, scale=1, size=10)
print(dog_whisker_length)
print(dog_ear_flapiness_index)
```

```python
cat_whisker_length = np.random.normal(loc=8, scale=1, size=10)
cat_ear_flapiness_index = np.random.normal(loc=5, scale=1, size=10)
print(cat_whisker_length)
print(cat_ear_flapiness_index)
```

```python
plt.scatter(dog_ear_flapiness_index, dog_whisker_length , label="dog")
plt.scatter(cat_ear_flapiness_index, cat_whisker_length, label="cat")
plt.xlabel("ear flappines")
plt.ylabel("whisker length")
plt.legend()
plt.show()
```

```python
def random_linear_classifier(data_dogs, data_cats, k, d):
  best_error =float('inf')
  best_tata = None
  best_tata0 = None

  for _ in range(k):
    tata = np.random.normal(size=d)
    tata0 = np.random.normal()

    error = compute_error(data_dogs, data_cats, tata, tata0)

    if error < best_error:
       best_error = error
       best_tata = tata
       best_tata0 = tata0

  return best_tata , best_tata0
```

```python
def compute_error(data_dogs, data_cats, tata, tata0):
    error = 0
    for dog_data in data_dogs:
        if np.dot(tata, dog_data)+tata0 <= 0:
           error += 1
    for cat_data in data_cats:
        if np.dot(tata, cat_data) + tata0 > 0:
           error += 1
    return error
```

```python
data_dogs = np.vstack((dog_whisker_length, dog_ear_flapiness_index)).T
data_cats = np.vstack((cat_whisker_length, cat_ear_flapiness_index)).T
```

```python
k = 100
d=2
best_tata, best_tata0 = random_linear_classifier(data_dogs, data_cats, k, d)
print(best_tata)
print(best_tata0)
```
